```python
search_cathegories = ['An Analysis', 'An Appraisal', 'Article', 'Blog', 'Brief', 'Caption', 'Chronology', 'Column', 'Correction', 'Editorial', 'Editorial Cartoon', 'First Chapter', 
                      'Front Page', 'Interview', 'List', 'News', 'News Analysis', 'Newsletter', 'Op-Ed', 'Postscript', 'Premium', 'Question', 'Recipe', 'Review', 'SectionFront', 
                      'Special Report', 'Summary', 'Text']
```

```python
def get_articles_for_year_and_page(year, api_key, page, search_cathegories):
    cathegories = 'type_of_material:("{}")'.format('" "'.join(search_cathegories))
    response = requests.get("https://api.nytimes.com/svc/search/v2/articlesearch.json?fq=pub_year:{} AND {}&api-key={}&page={}".format(year, cathegories, api_key, page))
    time.sleep(6)
    return response.json()

def get_articles_for_year(year, api_key, num_pages, search_cathegories):
    first = get_articles_for_year_and_page(year, api_key, 1, search_cathegories)
    print(first)
    for i in tqdm(range(num_pages - 1)):
        first['response']['docs'] += get_articles_for_year_and_page(year, api_key, i + 2, search_cathegories)['response']['docs']
    return first
def response_to_tuple(response):
    smp = response['response']['docs']
    pairs = list(map(lambda x: (x['pub_date'], x['web_url']), smp))
    return pairs
```

```python
for year in tqdm(range(1991, 2000)):
    response = get_articles_for_year(year, api_key, 100, search_cathegories)
    df = pd.DataFrame(response_to_tuple(response))
    df.to_csv("{}.csv".format(year))
```
